# 5.  Analyzing our data by answering some questions 

 # Which location have the fastest roller coasters?

```python
df.head()
```

```python
dr= df.query('Location ! = "Other"')\
.groupby('Location')['speed_mph']\
.agg(['mean', 'count'])\
.sort_values('mean', ascending = False)\
.query('count >= 10')

dr
```

```python
dr['mean'].plot(kind='barh').set(xlabel= 'Average Speed(mph)', title ='Avegare speed of Rolller Coasters by location')
```
